```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator, FuncFormatter, StrMethodFormatter, LinearLocator

plt.style.use('default')

x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)
y1 = 1 + np.exp(x)

fig, ax = plt.subplots(nrows=1, ncols=1)

ax.plot(x, y, linewidth=2, color='purple', label='sinus')
ax.plot([1, 2, 3, 4], [1, 5, 3, 7], marker='o', markersize=15, linewidth=3, markerfacecolor='green', markeredgecolor='blue', label='random data')

ax.set(xlabel='x', ylabel='y', title='Simple Plot', xlim=(0, 8), ylim=(0,8), xticks=np.arange(1, 8), yticks=np.arange(1, 8))
ax.grid(which='major', visible=True, color='g', linestyle=':')
ax.legend(['First Line', 'Second Line'], loc='best')
ax.yaxis.set_minor_locator(locator=AutoMinorLocator())
ax.xaxis.set_major_formatter(formatter=StrMethodFormatter('{x:.2f}'))
ax.spines['top'].set_visible(False)

ax.scatter

plt.show
```

```python
x = np.linspace(0, 5, 50)
fig, axs = plt.subplots(2, 2)
axs[0][0].plot(x, np.sin(x))
axs[0][1].plot(x, np.cos(x))
axs[1][0].plot(x, np.exp(x))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = axs[1][1].plot(t, s, lw=2)

axs[1][1].annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

axs[1][1].set_ylim(-2, 2)

plt.show()
```
